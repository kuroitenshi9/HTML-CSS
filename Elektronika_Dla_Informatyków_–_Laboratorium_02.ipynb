##1. Dioda Prostownicza




```
import pandas as pd
import matplotlib.pyplot as plt

'''
wz√≥r na rezystencje dynamiczna ‚ÑùùíÖùíöùíè = (ùëºùíå‚àí ùëºùíå‚àíùüè) / (ùë∞ùíå‚àí ùë∞ùíå‚àíùüè)
'''

U = list(range(0, 13))
I = [0, 0.58, 1.529, 2.504, 3.487, 4.474, 5.464, 6.455, 7.448, 8.441, 9.435, 10.43, 11.425]

Rdyn = [0]

for k in range(1, len(U)):
    delta_U = U[k] - U[k - 1]
    delta_I = I[k] - I[k - 1]
    R = delta_U / (delta_I / 1000)
    Rdyn.append(round(R,3))


wyniki = pd.DataFrame({
    "U [V]": U,
    "I [mA]": I,
    "Rdyn [Œ©]": Rdyn
})

fig, ax = plt.subplots(figsize=(12, 2.5))
ax.axis('off')
table_data = [
    ["U [V]"] + U,
    ["I [mA]"] + [f"{x:.3f}" for x in I],
    ["Rdyn [Œ©]"] + ["N/A" if x is None else str(x) for x in Rdyn]
]
table = ax.table(cellText=table_data, loc='center', cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)
plt.title("Tabela nr 1A: Pomiary diody prostowniczej w kierunku przewodzenia")


plt.figure(figsize=(8, 5))
plt.plot(U[0:], Rdyn[0:], marker='o', color='blue')
plt.title("Rezystancja dynamiczna dla diody prostowniczej w kierunku przewodzenia")
plt.xlabel("U [V]")
plt.ylabel("Rdyn [Œ©]")
plt.grid(True)
plt.tight_layout()
plt.xlim(0, max(U) if U else 1)
plt.ylim(0, max(Rdyn) if Rdyn else 1)
```
